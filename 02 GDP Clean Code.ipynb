# Опис на проект:

### Бруто-домашен производ (БДП)

- Бруто-домашен производ (БДП) (англиски: gross domestic product, GDP) — еден од основните показатели на големината и успехот на една економија. Бруто-домашниот производ е дефиниран како вкупната пазарна вредност на сите произведени финални добра и услуги во една земја во одреден период (вообичаено, една година). Според тоа, во БДП се вклучуваат само финалните добра, т.е. се исклучува вредноста на интермедијарите производи во производството. Изразот бруто се однесува на фактот дека амортизацијата на капиталните фондови е вклучена во пресметката на БДП. Значи, доколку се одземе амортизацијата на капиталните фондови би се добил нето домашен производ. Бруто-домашниот производ е централна макроекономска категорија преку која се мери вкупната економска активност на една земја. Во центарот на националните сметки е производството на стоките и услугите, односно бруто-домашниот производ (БДП), кој е главен показател за развиеноста на економијата на секоја држава. Со податокот за бруто-домашниот производ по жител се врши споредба на развиеноста на државата во однос на други држави, споредба на продуктивноста на трудот и истовремено податок од кој се добиваат податоците за бруто-националниот доход, бруто-штедењето и нето -задолжувањето на државата.

Бруто-домашен производ = Финална потрошувачка + бруто-инвестиции + извоз - увоз

### Финална потрошувачка

#### Трошоци за финална потрошувачка

- Трошоците за финална потрошувачка се составени од трошоци што ги прават резидентните институционални единици за стоки и за услуги што се употребуваат за директно задоволување на личните потреби или желби, како и на колективните потреби на членовите на заедницата. Трошоците за финална потрошувачка може да се направат на домашна територија или во странство.

Финална потрошувачка= Финална потрошувачка на домаќинствата + финална потрошувачка на НПИСД + финална потрошувачка на државата.

### Бруто-инвестиции

#### ИНВЕСТИЦИИ ВО ОСНОВНИ СРЕДСТВА

Инвестициите во основни средства се мерат преку вкупната вредност на набавените минус отуѓените основни средства од производителот во текот на пресметковниот период, плус некои унапредувања на вредноста на непроизведените средства реализирани преку производната активност на институционалната единица.

```python
df['Quarter'] = df['Unnamed: 0'].str[5:]
```

```python
df.drop(['Unnamed: 0'], axis = 1, inplace=True)
```

### Барање 3:

Најдете го Бруто-домашниот производ и зачувајте го во вчитаниот data frame како нова колона.

Бруто-домашен производ = Финална потрошувачка + бруто-инвестиции + извоз - увоз

```python
df['Gross_domestic_product'] = df['Final_public_spending'] + df['Gross_investment'] + df['Export_of_goods_and_services'] - df['Imports_of_goods_and_services']
df
```

### Барање 4:

Да се најде вкупната вредност на БДП од првите два квартали за 2017, 2018 и 2019 година. Потоа, да се визуелизираат тие со помош на pie chart и да се направи споредба измеѓу тие вредности.

```python
df['Year'].unique()
```

```python
df_year = df[(df['Year'] == 2017) | (df['Year'] == 2018) | (df['Year'] == 2019)]
df_year['Quarter'] = df_year['Quarter'].astype(int)
df_year
```

```python
df_quarter_1_2 = df_year[(df_year['Quarter'] == 1) | (df_year['Quarter'] == 2)]
```

```python
total = df_quarter_1_2.groupby(df_quarter_1_2['Year']).sum()
total.reset_index(inplace=True)
```

```python
plt.figure()
plt.pie(total['Gross_domestic_product'], autopct='%.0f%%')
plt.legend(total['Year'], bbox_to_anchor=(1.25, 1), loc='upper right', borderaxespad=0)
plt.show()
```

### Барање 5:

Да се направат boxplots за сите 4 квартали за 2017, 2018 и 2019 година за БДП за да се видат како се движеле вредностите за тие години. Што може да се заклучи од графикот?

---

```python
df_year
```

```python

# Sample data creation
np.random.seed(42)
df_year = {
    'Year': np.random.choice([2017, 2018, 2019], size=300),
    'Quarter': np.random.choice([1, 2, 3, 4], size=300),
    'Gross_domestic_product': np.random.randn(300)
}

# Create a boxplot using Seaborn catplot
sns.set(style="whitegrid")
plt.figure(figsize=(16, 10))

sns.boxplot(x="Year", y="Gross_domestic_product", hue="Quarter", data=df_year, palette="Set3", width=0.7)

# Adjust the plot labels and title
plt.title('Boxplot of Gross Domestic Product by Year and Quarter')
plt.xlabel('Year')
plt.ylabel('Gross Domestic Product')

# Show the plot
plt.show()
```

Conclusion: The biggest difference we can see in 2017 where the Gross_domestic_product in second and fourth quartal of the year is bigger then in the other years. Also we can see that the Gross_domestic_product is lowest in the 2019.

### Барање 8:

Да се најде вкупната финална потрошувачка за секоја година.

```python
df_final_spending = df.groupby('Year').sum()
df_final_spending.reset_index()
df_final_spending
```

```python
df_final_spending['Final_spending'] = df_final_spending['Final_household_spending_including_NPISD'] + df_final_spending['Final_public_spending']
df_final_spending.reset_index(inplace=True)
df_final_spending
```

```python
plt.figure()

plt.scatter(data=df_final_spending, x = 'Year', y = 'Final_spending')

plt.show()
```

Conclusion: The Final spending is increasing through the years.
